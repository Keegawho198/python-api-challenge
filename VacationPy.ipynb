```python
# Set parameters to search for a hotel
radius = 10000   #10,000 meters

latitude = hotel_df["Lat"]  # Replace with the actual latitude value
longitude = hotel_df["Lng"]  # Replace with the actual longitude value

params = {
    "apiKey": geoapify_key,  
    "categories": "accommodation.hotel",  # Filter to search for hotels
    "limit": 1,  # Limit the results to the first hotel found
    "lat": latitude,  # Add latitude
    "lon": longitude,  # Add longitude
    "filter": f"circle:{longitude},{latitude},10000"  # Define a circular area with a 10,000-meter radius
}
#categories=catering.restaurant

#response = requests.get(base_url, params=params).json()  # Make the request and convert the response to JSON format
#print(response)  # Debug: print the full response


# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude and longitude from the DataFrame
    latitude = row["Lat"]
    longitude = row["Lng"]

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Construct the params dictionary
    params = {
        "apiKey": geoapify_key,  # Your actual Geoapify API key
        "categories": "accommodation.hotel",  # Filter to search for hotels
        "limit": 1,  # Limit the results to the first hotel found
        "lat": latitude,  # Add latitude
        "lon": longitude,  # Add longitude
        "filter": f"circle:{longitude},{latitude},10000"  # Define a circular area with a 10,000-meter radius
    }

    # Make an API request using the params dictionary
    response = requests.get(base_url, params=params).json()  # Make the request and convert the response to JSON format

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = response["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results in the desired format
    print(f"{row['City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```
